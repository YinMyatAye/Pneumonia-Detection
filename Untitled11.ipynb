```python
import pathlib

import numpy as np
import pandas as pd

from sklearn.model_selection import train_test_split


import cv2
import PIL
import seaborn as sns
import matplotlib.pyplot as plt

import tensorflow as tf
import tensorflow.keras.backend as K
from tensorflow.keras.preprocessing.image import ImageDataGenerator
from tensorflow.keras.layers import Conv2D,Flatten,Dense,RandomFlip,RandomZoom,RandomRotation,Dropout,MaxPooling2D,Input
from tensorflow.keras.models import Sequential
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.math import confusion_matrix
```

```python
data_dir=pathlib.Path(khoongweihao_covid19_xray_dataset_train_test_sets_path) / 'xray_dataset_covid19'
data_dir
```

```python
train_normal=data_dir / 'train' / 'NORMAL'
train_pneumonia=data_dir / 'train' / 'PNEUMONIA'

test_normal=data_dir / 'test' / 'NORMAL'
test_pneumonia=data_dir / 'test' / 'PNEUMONIA'
```

```python
train_images_dict={
    'normal':list(train_normal.glob('*.jpeg')),
    'pneumonia':list(train_pneumonia.glob('*.jpeg'))
}

test_images_dict={
    'normal':list(test_normal.glob('*.jpeg')),
    'pneumonia':list(test_pneumonia.glob('*.jpeg'))
}

labels_dict={
    'normal':0,
    'pneumonia':1
}

print("Train Images Dictionary:", train_images_dict)
print("Test Images Dictionary:", test_images_dict)
```

```python
X,y=[],[]
for label,images in train_images_dict.items():
    for image in images:
        img=cv2.imread(str(image),0)
        img=cv2.resize(img,(200,200),interpolation=cv2.INTER_AREA)
        X.append(img)
        y.append(labels_dict[label])

X=np.array(X)
X=X/255
y=np.array(y)

X_test,y_test=[],[]
for label,images in test_images_dict.items():
    for image in images:
        img=cv2.imread(str(image),0)
        img=cv2.resize(img,(200,200),interpolation=cv2.INTER_AREA)
        X_test.append(img)
        y_test.append(labels_dict[label])

X_test=np.array(X_test)
X_test=X_test/255
y_test=np.array(y_test)
```

```python
X_train,X_val,y_train,y_val=train_test_split(X,y,stratify=y,random_state=69,test_size=0.25)
```

```python
import os
print(os.listdir(data_dir))
```

```python
X_train=np.expand_dims(X_train,-1)
X_test=np.expand_dims(X_test,-1)
X_val=np.expand_dims(X_val,-1)
```

```python
datagen=ImageDataGenerator(
    rotation_range=20,
    width_shift_range=0.3,
    height_shift_range=0.3,
    shear_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest',
)
```

```python
model=Sequential([
    Conv2D(filters=32,kernel_size=4,activation='relu'),
    MaxPooling2D(),

    Conv2D(filters=32,kernel_size=3,activation='relu'),
    MaxPooling2D(),

    Flatten(),
    Dropout(0.5),

    Dense(128,activation='relu'),
    Dropout(0.5),

    Dense(1,activation='sigmoid')

])
```

```python
model.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
X_train.shape
```

```python
earlystopping=tf.keras.callbacks.EarlyStopping(
    patience=3,
)
```

```python
model.fit(
    datagen.flow(X_train,y_train,batch_size=32),
    validation_data=datagen.flow(X_val,y_val,batch_size=16),
    shuffle=True,
    epochs=50,
    callbacks=[earlystopping]
)
```

```python
losses = pd.DataFrame(model.history.history)
losses[['accuracy','val_accuracy']].plot()
```

```python
model.evaluate(X_test,y_test)
```

```python
cf_matrix=confusion_matrix(y_test,model.predict(X_test)>0.5)
```

```python
def plot_cf(cf_matrix):
    ax = sns.heatmap(cf_matrix, annot=True, cmap='Blues')

    ax.set_title('Seaborn Confusion Matrix with labels\n\n');
    ax.set_xlabel('\nPredicted Values')
    ax.set_ylabel('Actual Values ');
    ax.xaxis.set_ticklabels(['0','1'])
    ax.yaxis.set_ticklabels(['0','1'])
    plt.show()
```
